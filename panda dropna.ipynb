# Group By Function

```python
import pandas as pd
```

```python
Data = {'name':['vishesh','rohan','rahul','tarun','nakul','nishant'],
        'age':[19,15,13,16,14,17],
        'marks':[80,70,60,40,30,50],
        'class':[12,10,7,9,11,5]}
Data = pd.DataFrame(Data)
```

```python
Data
```

```python
name=Data.groupby('name')['marks'].sum()
name
```

```python
age=Data.groupby('age')['class'].sum()
age
```

```python
a=Data.groupby('name').agg({'marks':['sum','mean','count']})
a
```
